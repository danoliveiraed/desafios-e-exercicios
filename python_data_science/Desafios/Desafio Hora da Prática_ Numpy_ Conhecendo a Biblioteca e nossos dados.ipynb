# Coeficiente linear toranjas e laranjas

```python
#Laranjas
X = laranjas_peso
Y = laranjas_diametro
n = np.size(Y)
```

a = [n * Σ(XiYi) - ΣXi * ΣYi] / [n * Σ(Xi²) - (ΣXi)²]

```python
a = (n * np.sum(X * Y) - np.sum(X) * np.sum(Y)) / (n * np.sum(X**2) - np.sum(X)**2)
```

```python
print(a)
```

b = média(Y) - a * média(X)

```python
b = np.mean(Y) - a * np.mean(X)
print(b)
```

Cálculo da reta

```python
y = a * X + b
print(y)
```

```python
np.linalg.norm(Y - y)
```

```python
plt.plot(laranjas_peso, laranjas_diametro)
plt.plot(laranjas_peso, y)
```

```python
#Toranjas
X = toranjas_peso
Y = toranjas_diametro
n = np.size(Y)
```

```python
a = (n * np.sum(X * Y) - np.sum(X) * np.sum(Y)) / (n * np.sum(X**2) - np.sum(X)**2)
```

```python
b = np.mean(Y) - a * np.mean(X)
print(b)
```

```python
np.linalg.norm(Y - y)
```
